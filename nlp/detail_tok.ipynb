```python
import pandas as pd
import numpy as np
from konlpy.tag import Komoran
from konlpy.tag import Okt
import os
```

```python
okt = Okt()

def tokenizer_okt_morphs(doc):
    return okt.morphs(doc)
def tokenizer_okt_nouns(doc):
    return okt.nouns(doc)
def tokenizer_okt_pos(doc):
    return okt.pos(doc)
```

```python
df = pd.read_csv('detailinfo.csv')
```

```python
print(df.columns)
```

```python
df = df.loc[:,['bookname','isbn13', 'description']]
df.head()
```

```python
df.fillna('없음', inplace=True)
```

```python
df.info()
```

```python
df['tokenized'] = df['description'].apply(tokenizer_okt_nouns)
```

```python
df.head()
```

```python
df = df.loc[:,['bookname','isbn13','tokenized']]
```

```python
df.to_csv('detail_tokenized.csv',encoding='utf-8')
```

```python
vocab_size=50
tokenizer = Tokenizer(num_words=vocab_size+1)
tokenizer.fit_on_texts(text)
```

```python
print(tokenizer.word_index)
```

```python
intEnc = tokenizer.texts_to_sequences(text)
print(intEnc)
```

```python
type(intEnc)
```

```python
max_len = max(len(i) for i in intEnc)
max_len
```

```python
for i in intEnc:
    while len(i) < max_len:
        i.append(0)
padded_np = np.array(intEnc)
padded_np
```

```python
p = padded_np.tolist()
pp = pd.Series(p)
df['encoding'] = pp
df = df.iloc[1:,]
df.to_csv('detailinfo_nlp_preprocessing.csv',encoding='utf-8',index=False)
```

```python
from sklearn.feature_extraction.text import TfidfVectorize
```

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['tokenized'])
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(df.index, index = df['bookname']).drop_duplicates()
indices
```

```python
def get_recommandations(bookname,cosine_sim=cosine_sim):
    idx = indices[bookname]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x:x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    books_indices = [i[0] for i in sim_scores]
    return df[['bookname','isbn13']].iloc[books_indices]
```
